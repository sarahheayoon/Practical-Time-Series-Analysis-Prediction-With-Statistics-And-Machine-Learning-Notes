```r
install.packages("zoo")
install.packages("data.table")
```

```r
library(zoo)
library(data.table)
library(tidyverse)
library(lubridate)
```

```r
unemp <- read.csv("UNRATE.csv")
```

```r
head(unemp,5)
```

```r
unemp %>%
    mutate(year = year(DATE)) %>%
    group_by(year) %>%
    summarise(yearly_rate = mean(UNRATE))%>%
    filter(year>1977, year<1982)%>%
    ggplot(aes(x = year, y = yearly_rate))+
    geom_line()
```

```r
unemp %>% 
 mutate(year = year(DATE), month = month(DATE)) %>% 
 group_by(year) %>% 
 summarise(yearly = mean(UNRATE))%>%
 filter(year>2000, year<2010)%>%
 ggplot(aes(x=year, y=yearly)) +
 geom_line()
```
